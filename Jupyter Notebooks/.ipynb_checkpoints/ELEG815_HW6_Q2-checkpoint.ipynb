```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
path = 'D:\Courses\Fall 19\ELEG 815 Statistical Learning\HW6\Dataset'
train_data = pd.read_csv(path+'\DigitsTraining.csv',header=None)
test_data = pd.read_csv(path+'\DigitsTesting.csv',header=None)
```

```python
def Get_Two_Digits(data,digit_1,digit_2):
    dataset = pd.DataFrame()
    for i in range(len(data)):
        if int(data.iloc[i][0]) == digit_1 or int(data.iloc[i][0]) == digit_2:
            # https://stackoverflow.com/questions/16597265/appending-to-an-empty-dataframe-in-pandas
            """Pandas doesnt append in place. Have to store in a new variable"""
            dataset = dataset.append(data.iloc[i],ignore_index=True)
    return dataset
raw_training_dataset = Get_Two_Digits(train_data,1,2)
raw_testing_dataset = Get_Two_Digits(test_data,1,2)
```

```python
def Extract_features(raw_dataset,digit_1,digit_2):
    len_dataset = len(raw_dataset)
    number_of_features = 2
    dataset = np.empty((len_dataset,number_of_features + 1))
    true_labels = np.empty((len_dataset,))
    for index,image_pixels in raw_dataset.iterrows():
        """---------------------- Average Intensity-------------------------"""
        avg_intensity = sum(image_pixels[1:])/len(image_pixels[1:])

        """-----------------------Average Symmetry--------------------------"""
        img = np.array(image_pixels[1:]).reshape((16,16))
        flipped_image = np.fliplr(img)
        flipped_image_pixels = flipped_image.flatten()
        image_pixels = np.array(image_pixels)
        asymmetry = sum(np.absolute(image_pixels[1:] - flipped_image_pixels))/len(image_pixels[1:])
        symmetry = 1 - asymmetry
        dataset[index][0] = 1
        dataset[index][1] = 1 - avg_intensity           # Higher value of intensity means there are more white pixels than black pixels because white = 255 and black = 0
        dataset[index][2] = symmetry
        if int(image_pixels[0]) == digit_1:                   # Suppose digit 2 has label = -1
            true_labels[index] = -1
        else:
            true_labels[index] = 1                      # Suppose digit 6 has label +1
    return dataset,true_labels
```

```python
def Non_Linear_Transform(existing_dataset,existing_labels):
    #new_feature = np.empty((len(existing_dataset),7))
    new_dataset = np.empty((len(existing_dataset),existing_dataset.shape[1]+7))
    new_dataset[:,:3] = existing_dataset
    for i in range(len(existing_dataset)):
        #new_feature[i] = existing_dataset[i][2]**2
        new_dataset[i][3] = existing_dataset[i][1]**2
        new_dataset[i][4] = existing_dataset[i][1]*existing_dataset[i][2]
        new_dataset[i][5] = existing_dataset[i][2]**2
        new_dataset[i][6] = existing_dataset[i][1]**3
        new_dataset[i][7] = existing_dataset[i][1]**2 * existing_dataset[i][2]
        new_dataset[i][8] = existing_dataset[i][1] * existing_dataset[i][2]**2
        new_dataset[i][9] = existing_dataset[i][2]**3
    return new_dataset,existing_labels
```

```python
def Set_Colors_Array(true_labels):
    col = [None]*len(true_labels)
    for i in range(len(true_labels)):
        if int(true_labels[i]) == -1:
            col[i] = 'r'
        else:
            col[i] = 'g'
    return col
def Scatter_Plot(dataset,colors):           # Red = digit 2 and Green = digit 1
    x = dataset[:,1]
    y = dataset[:,2]
    plt.scatter(x,y,c= colors,marker='.')
    plt.xlabel('Average Intensity')
    plt.ylabel('Average Symmetry')
    plt.show()
```

```python
def Linear_Regression_Original_Data(X,Y):
    XTranspose_X_Inv = np.dot(np.linalg.inv(np.dot(X.T,X)),X.T)
    return np.dot(XTranspose_X_Inv,Y)
```

```python
def Plot_Results(dataset,colors,W,digit):
    x = dataset[:,1]
    y = dataset[:,2]
    plt.scatter(x,y,c= colors,marker='.')
    plt.xlabel('Average Intensity')
    plt.ylabel('Average Symmetry')
    plt.title('Linear Regression Boundary for Digits 1 and '+str(digit))
    X = dataset
    Y = [0]*len(X)
    for i in range(len(X)):
       Y[i] = (-W[0]*X[i][0]- W[1]*X[i][1])/W[2]
    plt.plot(x,Y)
```

```python
def Error(dataset,true_labels,W):
    count = 0
    for index,point in enumerate(dataset):
        if np.sign(np.dot(W.T,point)) != true_labels[index]:
            count += 1
    return (count/len(dataset))*100
#print('Error of Training Data :',Error(train_dataset,train_true_labels,W))
```

```python
digits = [0,2,3,4,5,6,7,8,9]
def Separate_All_Digits(digits,train_data,test_data):
    for index,digit in enumerate(digits):
        print('Generating Boundary for digits : 1 and ',str(digit) +' ...')
        raw_training_dataset = Get_Two_Digits(train_data,1,digit)
        raw_testing_dataset = Get_Two_Digits(test_data,1,digit)
        d_3_train_data,d_3_train_true_labels = Extract_features(raw_training_dataset,1,digit)
        d_3_test_data,d_3_test_true_labels = Extract_features(raw_testing_dataset,1,digit)
        colors_train = Set_Colors_Array(d_3_train_true_labels)
        colors_test = Set_Colors_Array(d_3_test_true_labels)
        W_d_3 = Linear_Regression_Original_Data(d_3_train_data,d_3_train_true_labels)
        E_in = Error(d_3_train_data,d_3_train_true_labels,W_d_3)
        E_out = Error(d_3_test_data,d_3_test_true_labels,W_d_3)
        print('E_in for digits : 1 and '+str(digit)+' is '+ str(E_in))
        print('E_out for digits : 1 and '+str(digit)+' is '+ str(E_out))
        plt.figure(index)
        Plot_Results(d_3_train_data,colors_train,W_d_3,digit)
Separate_All_Digits(digits,train_data,test_data)
```

```python
def Separate_All_Digits_Transformed(digits,train_data,test_data):
    for index,digit in enumerate(digits):
        print('Generating Boundary for digits : 1 and ',str(digit) +' ...')
        raw_training_dataset = Get_Two_Digits(train_data,1,digit)
        raw_testing_dataset = Get_Two_Digits(test_data,1,digit)
        d_3_train_data,d_3_train_true_labels = Extract_features(raw_training_dataset,1,digit)
        d_3_test_data,d_3_test_true_labels = Extract_features(raw_testing_dataset,1,digit)
        d_10_train_data,d_10_train_true_labels = Non_Linear_Transform(d_3_train_data,d_3_train_true_labels)
        d_10_test_data,d_10_test_true_labels = Non_Linear_Transform(d_3_test_data,d_3_test_true_labels)
        colors_train = Set_Colors_Array(d_10_train_true_labels)
        colors_test = Set_Colors_Array(d_10_test_true_labels)
        W_d_10 = Linear_Regression_Original_Data(d_10_train_data,d_10_train_true_labels)
        #print('Weight vector for digits : 1 and '+str(digit)+' is :'+str(W_d_10))
        E_in = Error(d_10_train_data,d_10_train_true_labels,W_d_10)
        E_out = Error(d_10_test_data,d_10_test_true_labels,W_d_10)
        print('E_in for digits : 1 and '+str(digit)+' is '+ str(E_in))
        print('E_out for digits : 1 and '+str(digit)+' is '+ str(E_out))
        plt.figure(index)
        #Plot_Results(d_3_train_data,colors_train,W_d_10,digit)
Separate_All_Digits_Transformed(digits,train_data,test_data)
```
